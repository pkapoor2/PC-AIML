### get missing values

```python
# find missing values
display(df.isnull().sum())
# find missing values in percentage
display(df.isnull().mean().round(4) * 100)
# find missing values in percentage for numeric columns
display(df.select_dtypes(include=[np.number]).isnull().mean().round(4) * 100)
# find missing values in percentage for categorical columns
display(df.select_dtypes(exclude=[np.number]).isnull().mean().round(4) * 100)
```

```python
df.describe()
```

```python
# create a new dataframe with missing values
df_missing = df.copy(deep=True)
# go to each row and add nan to a random colum with probability 0.01
for i in range(len(df_missing)):
    if np.random.rand() < 0.05:  # 1% chance
        col = np.random.choice(df_missing.columns)
        df_missing.at[i, col] = np.nan
# find missing values in the new dataframe
display(df_missing.isnull().sum())

```

### imputaions

```python
#visualise missing values to choose an imputation strategy
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_missing.isnull(), cbar=False, cmap='viridis')

```

```python
#save rows with missing values to a new dataframe
df_missing_rows = df_missing[df_missing.isnull().any(axis=1)]
```

```python
# sqft,yr_built
df_impute = df_missing.copy(deep=True)
#impute missing values with mean for numeric columns
for col in df_impute.select_dtypes(include=[np.number]).columns:
    df_impute.fillna({col: df_impute[col].mean()}, inplace=True)
# impute missing values with mode for categorical columns
for col in df_missing.select_dtypes(exclude=[np.number]).columns:
    df_impute.fillna({col: df_impute[col].mode()[0]}, inplace=True)
# check if there are still missing values
display(df_impute.isnull().sum())
# check if the imputation worked
display(df_impute.describe(include='all').round(1))
sns.heatmap(df_impute.isnull(), cbar=False, cmap='viridis')
```

```python
#compare stats of df for missing values before and after imputation
display(df_missing.describe(include='all').round(1))
# compare stats of df for missing values before and after imputation
display(df_impute.describe(include='all').round(1))


```

### data cleaning

```python
#clean df
df_clean = df.copy(deep=True)
df_clean.describe()
```

```python
df_clean['date'] = pd.to_datetime(df_clean['date'])
# convert sqft_living to numeric
df_clean['sqft_living'] = pd.to_numeric(df_clean['sqft_living'], errors='coerce')
# convert sqft_lot to numeric
df_clean['sqft_lot'] = pd.to_numeric(df_clean['sqft_lot'], errors='coerce')
# convert yr_built to numeric
df_clean['yr_built'] = pd.to_numeric(df_clean['yr_built'], errors='coerce')
# convert sqft_above to numeric
df_clean['sqft_above'] = pd.to_numeric(df_clean['sqft_above'], errors='coerce')
# convert sqft_basement to numeric
df_clean['sqft_basement'] = pd.to_numeric(df_clean['sqft_basement'], errors='coerce')
df_clean.describe()
```

### log transform

```python
#identify numeric columns to tranform
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
#which columns to log transform based on skewness
skewed_cols = df_clean[numeric_cols].apply(lambda x: abs(x.skew())).sort_values(ascending=False)
# log transform skewed columns
display(df_clean[numeric_cols].apply(lambda x: x.skew()).sort_values(ascending=False))
# apply log transformation to skewed columns
for col in skewed_cols[skewed_cols > 0.5].index:
    df_clean[col] = np.log1p(df_clean[col])
# check the skewness after transformation
display(df_clean[numeric_cols].apply(lambda x: x.skew()).sort_values(ascending=False))


```

```python
#visualise orignal vs windsorized price to illustrate the effect of outlier removal
sns.histplot(df_clean['price'], kde=True, color='blue', label='Original Price', stat='density')
sns.histplot(df_clean['price'].clip(lower=df_clean['price'].quantile(0.01), upper=df_clean['price'].quantile(0.99)), 
             kde=True, color='red', label='Windsorized Price', stat='density')
plt.legend()
# remove outliers using windsorization
df_clean['price'] = df_clean['price'].clip(lower=df_clean['price'].quantile(0.01), upper=df_clean['price'].quantile(0.99))
# check the skewness after outlier removal
display(df_clean['price'].skew())
#visualise the price distribution after outlier removal
sns.histplot(df_clean['price'], kde=True, color='red', label='Windsorized Price', stat='density')
```

```python
#example concatinating two dataframes
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})
# concatenate dataframes
df_concat = pd.concat([df1, df2], ignore_index=True)
# display the concatenated dataframe
display(df_concat)
#example merging two dataframes
df3 = pd.DataFrame({'A': [1, 2, 3], 'C': [4, 5, 6]})
df4 = pd.DataFrame({'A': [2, 3, 4], 'D': [7, 8, 9]})
# merge dataframes on column A
df_merge = pd.merge(df3, df4, on='A', how='inner')
# display the merged dataframe
display(df_merge)
# merge dataframes on column A outer
df_merge = pd.merge(df3, df4, on='A', how='outer')
# display the merged dataframe
display(df_merge)
```

```python
#reshaoing dataframes
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
display(df)
# melt the dataframe
df_melted = pd.melt(df, id_vars=['A'], value_vars=['B', 'C'], var_name='Variable', value_name='Value')
# display the melted dataframe
display(df_melted)
# pivot the melted dataframe
df_pivoted = df_melted.pivot(index='A', columns='Variable', values='Value').reset_index()
# display the pivoted dataframe
display(df_pivoted)
# example of stacking and unstacking
df_stack = df.set_index(['A', 'B']).stack().reset_index(name='Value')
# display the stacked dataframe
display(df_stack)
# unstack the stacked dataframe
df_unstack = df_stack.set_index(['A', 'B']).unstack().reset_index()
# display the unstacked dataframe
display(df_unstack)
```
