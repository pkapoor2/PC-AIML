```python
import pandas as pd

#name age salary dict for employee
data_dict = {
    'name': ['John', 'Anna', 'Peter', 'Linda'],
    'age': [28, 24, 35, 32],
    'salary': [50000, 60000, 70000, 80000]
    }
dfDict = pd.DataFrame(data_dict)
print(dfDict)
```

```python
import numpy as np
#name age salary dict for employee
npArray = np.array([
    ['John', 28, 50000],
    ['Anna', 24, 60000],
    ['Peter', 35, 70000],
    ['Linda', 32, 80000]
    ])
dfArray = pd.DataFrame(npArray, columns=['name', 'age', 'salary'])
print(dfArray)
```

```python
#dfDict to csv file
dfDict.to_csv('employee.csv', index=False)
#new rows in dfArray not in dfDict
dfArray = dfArray[~dfArray.isin(dfDict)].dropna()
#dfArray to csv file

dfArray.to_csv('employee.csv', index=False, mode='a', header=False)
#read csv file
df = pd.read_csv('employee.csv')
print(df)
```
