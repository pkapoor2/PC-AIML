- Create dummy variables for categorical variables

```python
def hot_encode(df, catagoricalFeatures):
    # Create dummy variables for each category in the column
    dummies_df = pd.get_dummies(df[catagoricalFeatures], prefix=catagoricalFeatures).astype(int)
    # Concatenate the original dataframe with the new dummy variables
    df = pd.concat([df, dummies_df], axis=1)
    # Drop the original categorical column
    df.drop(catagoricalFeatures, axis=1, inplace=True)
    return df
    
hot_encoded_df = hot_encode(df, catagoricalFeatures)
hot_encoded_df
```

```python


#hot_encoded_df.drop(columns=['Unnamed: 0'], inplace=True)
hot_encoded_df.drop(columns=['dateSold'], inplace=True)
```
