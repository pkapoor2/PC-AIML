## Task 2: Import the CSV file – NSMES1988new.csv into a dataframe.

```python
df = pd.read_csv('../dataset/NSMES1988.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

## Task 4: Perform the following operations on age and income columns. Multiply age by 10 and income by 10000.

```python
df['age'] = df['age']*10
df['income'] = df['income']*10000
df['age'] = df['age'].astype(np.int16)
df['income'] = df['income'].astype(np.int64)
df.info(memory_usage='deep')
df[['age','income']].head()
```

## Task 5: Perform basic statistical analysis on the new dataframe and generate a brief report on the outcome. Save the dataframe as NSMES1988updated.csv file in the local space for possible future use.

```python
df_optimized = df.copy()
objects = df_optimized.select_dtypes(include=['object']).columns
print("listing object columns values to see if this is a good candidate for conversion")
candidate_objects = [col for col in objects if len(df_optimized[col].unique()) < 0.5*len(df_optimized)]
# Convert object columns to category
for col in candidate_objects:
    df_optimized[col] = df_optimized[col].astype('category')
```

```python
#downcast the numeric columns
numeric_cols = df_optimized.select_dtypes(include=['int', 'float']).columns
for col in numeric_cols:
    if df_optimized[col].dtype == 'int64':
        df_optimized[col] = pd.to_numeric(df_optimized[col], downcast='integer')
    elif df_optimized[col].dtype == 'float64':
        df_optimized[col] = pd.to_numeric(df_optimized[col], downcast='float')
# Check memory usage after optimization
```

```python
def mem_usage(df, info=False):
    if(info):
        print("Memory usage of the DataFrame:")
        print(df.info(memory_usage='deep'))
        print(df.memory_usage(deep=True))
    return df.memory_usage(deep=True).sum() / 1024**2
        
print(f'before: {mem_usage(df): .2f} vs after: {mem_usage(df_optimized): .2f}')
print(f'Dataframe size reduced by: {((mem_usage(df)-mem_usage(df_optimized))*100)/mem_usage(df): .2f} percentage')
```
