```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli

# Bernoulli distribution parameter
p = 1 /6 # Probability of success

# Theoretical probabilities
x = [0, 1]
probs = [1 - p, p]

# Plot the distribution as a bar chart
plt.bar(x, probs, width=0.4, color='pink', edgecolor='black')

# Add labels and title
plt.xticks([0, 1])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution (p = 1/6)')

# Show the plot
plt.show()
```

```python
#binomial distribution
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Generate empirical data
n=10
p=0.5
data = binom.rvs(n=n, p=p, size=1000)

# Histogram: empirical probabilities
counts, bins = np.histogram(data, bins=np.arange(n+2), density=True)
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

# Plot both
plt.bar(x - 0.2, counts, width=0.4, label='Empirical', color='lightgreen', edgecolor='black')
plt.bar(x + 0.2, pmf, width=0.4, label='Theoretical PMF', color='green', edgecolor='black')

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution Comparison (n={n}, p={p})')
plt.legend()
plt.show()

```

Poison Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Generate Poisson-distributed data
mu = 4
data_poisson = poisson.rvs(mu, size=10000)

# Plot histogram with discrete bins centered on integer values
bins = np.arange(data_poisson.min(), data_poisson.max() + 1.5) - 0.5
plt.hist(data_poisson, bins=bins, density=True, alpha=0.6, color='lightgreen', edgecolor='black')

# Labels
plt.xlabel('Number of Events (k)')
plt.ylabel('Relative Frequency')
plt.title('Empirical Histogram of Poisson Distribution (μ = 4)')

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Scipy.stats poisson class is used along with PMF method to calculate the value of probabilities
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Random variable representing number of restaurants
# Mean number of occurrences of restaurants in 10 miles is 2
X = [0, 1, 2, 3, 4, 5]
lmbda = 2

# Probability values
poisson_pd = poisson.pmf(X, lmbda)

# Plot the probability distribution
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(X, poisson_pd, 'bo', ms=8, label='poisson pmf')
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X = No. of restaurants", fontsize="18")
plt.title("Poisson distribution - No. of restaurants vs probability", fontsize="18")
ax.vlines(X, 0, poisson_pd, colors='b', lw=5, alpha=0.5)
```

normal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, gaussian_kde

# Generate normally distributed data
mu = 0      # Mean
sigma = 1   # Standard deviation
data_normal = norm.rvs(mu, sigma, size=10000)

# Plot histogram
plt.hist(data_normal, bins=30, density=True, alpha=0.5, color='blue')

# Calculate the KDE
kde = gaussian_kde(data_normal)
kde_xs = np.linspace(data_normal.min(), data_normal.max(), 301)
kde_ys = kde.pdf(kde_xs)

# Plot KDE
plt.plot(kde_xs, kde_ys, color='red')

# Add labels
plt.xlabel('Normal distribution')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

Uniform Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

# Generate uniformly distributed data
a, b = 0, 10  # Define the start and end points of the uniform distribution
data_uniform = uniform.rvs(a, b-a, size=10000)

# Plot histogram
plt.hist(data_uniform, bins=30, density=True, alpha=0.5, color='blue')

# Calculate the PDF of the Uniform Distribution
uniform_xs = np.linspace(a, b, 301)
uniform_ys = uniform.pdf(uniform_xs, a, b-a)

# Plot PDF
plt.plot(uniform_xs, uniform_ys, color='red')

# Add labels
plt.xlabel('Uniform Distribution')
plt.ylabel('Density')

# Display the plot
plt.show()
```

cdf of normal Distribution

```python
# CDF Example 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate x values
x = np.linspace(-4, 4, 500)

# Compute the CDF of the standard normal distribution
cdf = norm.cdf(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x, cdf, color='blue', linewidth=2)
plt.title('CDF of Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('F(x) = P(X ≤ x)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

cdf of discreate numbers

```python
# CDF Example 2
import matplotlib.pyplot as plt
import numpy as np

# Define discrete values and their probabilities
x = [-1, 0, 1]
cdf = [0, 0.5, 1.0]  # cumulative probabilities

# Prepare step plot for CDF
plt.figure(figsize=(8, 4))
plt.step([-1,0, 1], [0, 0.5, 1.0], where='post', label='CDF', color='purple')
plt.hlines(0, -0.5, 0, linestyles='dashed', colors='gray')
plt.hlines(0.5, 0, 1, linestyles='dashed', colors='gray')
plt.hlines(1.0, 1, 1.5, linestyles='dashed', colors='gray')

# Plot markers for jumps
plt.plot(x, cdf, 'o', color='purple')

# Axis formatting
plt.xticks([-1, 0, 1])
plt.xlim(-1.5, 1.5)
plt.ylim(0, 1.1)
plt.xlabel('x')
plt.ylabel('F(x) = P(X ≤ x)')
plt.title('CDF of a Discrete Random Variable: X ∈ {0, 1}, P=0.5')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
