```python
#class to lear data visualization of pandas dataframe
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# generate a sinewave dataframe for testing line graph
def generate_sinewave_df(num_points=1000, frequency=1, amplitude=1, phase=0):
    x = np.linspace(0, 2 * np.pi * frequency, num_points)
    y = amplitude * np.sin(x + phase)
    df = pd.DataFrame({'x': x, 'y': y})
    return df
# plot sine wave
def plot_sinewave(df):
    plt.figure(figsize=(10, 5))
    plt.plot(df['x'], df['y'], label='Sine Wave', color='blue')
    plt.title('Sine Wave')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid()
    plt.show()
plot_sinewave(generate_sinewave_df())
```

```python
# generate a random data set for testing scatter plot
def generate_random_df(num_points=1000, allignment=0.1):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    x = np.random.rand(num_points) * 100
    y = np.random.rand(num_points) * 100
    # Add some alignment to the data points
    y += allignment * x + np.random.normal(0, 10, num_points)  # Adding noise
    df = pd.DataFrame({'x': x, 'y': y})
    return df
# plot scatter plot
def plot_scatter(df):
    plt.figure(figsize=(10, 5))
    plt.scatter(df['x'], df['y'], label='Random Points', color='red', alpha=0.5)
    plt.title('Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid()
    plt.show()
plot_scatter(generate_random_df())
```

```python
# generate a random data set for testing bar plot
def generate_bar_df(num_categories=10, num_points=1000):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    categories = [f'Category {i}' for i in range(num_categories)]
    values = np.random.randint(1, 100, size=num_categories)
    df = pd.DataFrame({'Category': categories, 'Value': values})
    return df
# plot bar plot
def plot_bar(df):
    plt.figure(figsize=(10, 5))
    sns.barplot(x='Category', y='Value', data=df)
    plt.title('Bar Plot')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()
plot_bar(generate_bar_df())
```

```python
# generate a random data set for testing box plot
def generate_box_df(num_categories=10, num_points=1000):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    categories = [f'Category {i}' for i in range(num_categories)]
    values = np.random.randint(1, 100, size=(num_points, num_categories))
    df = pd.DataFrame(values, columns=categories)
    return df
# plot box plot
def plot_box(df):
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=df, orient='h')
    plt.title('Box Plot')
    plt.xlabel('Values')
    plt.ylabel('Categories')
    plt.grid(axis='x')
    plt.show()
plot_box(generate_box_df())
```

```python
# generate a random data set for testing radar plot
def generate_radar_df(num_categories=10):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    categories = [f'Category {i}' for i in range(num_categories)]
    values = np.random.randint(1, 100, size=num_categories)
    df = pd.DataFrame({'Category': categories, 'Value': values})
    return df
# plot radar plot
def plot_radar(df):
    # Number of variables
    num_vars = len(df['Category'])

    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # The plot is a circle, so we need to "complete the loop" and append the start value to the end.
    values = df['Value'].values.flatten().tolist()
    values += values[:1]
    angles += angles[:1]

    # Draw the radar chart
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.fill(angles, values, color='red', alpha=0.25)
    ax.plot(angles, values, color='red', linewidth=2)

    # Labels for each axis
    plt.xticks(angles[:-1], df['Category'], color='black', size=10)
    plt.title('Radar Plot')
    plt.show()
plot_radar(generate_radar_df())


```

```python
# generate a random data set for testing area plot
def generate_area_df(num_points=1000):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    x = np.linspace(0, 10, num_points)
    y1 = np.sin(x) + 1
    y2 = np.cos(x) + 1
    df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})
    return df
# plot area plot
def plot_area(df):
    plt.figure(figsize=(10, 5))
    plt.fill_between(df['x'], df['y1'], color='blue', alpha=0.5, label='Area 1')
    plt.fill_between(df['x'], df['y2'], color='red', alpha=0.5, label='Area 2')
    plt.title('Area Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid()
    plt.show()
plot_area(generate_area_df())
```

```python
#generate a random data set for testing word cloud
def generate_wordcloud_df(num_words=100):
    # Generate random words and their frequencies
    np.random.seed(0)  # For reproducibility
    words = [f'Word {i}' for i in range(num_words)]
    frequencies = np.random.randint(1, 100, size=num_words)
    df = pd.DataFrame({'Word': words, 'Frequency': frequencies})
    return df
# plot word cloud
def plot_wordcloud(df):
    #install wordcloud library in this environment
    try:
        from wordcloud import WordCloud
    except ImportError:
        !pip install wordcloud
        from wordcloud import WordCloud
        return
    

    # Create a dictionary of words and their frequencies
    word_freq = dict(zip(df['Word'], df['Frequency']))

    # Generate the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud')
    plt.show()
plot_wordcloud(generate_wordcloud_df())
```

```python
# generate a random data set for testing 3d scatter plot
def generate_3d_scatter_df(num_points=1000):
    # Generate random data points
    np.random.seed(0)  # For reproducibility
    x = np.random.rand(num_points) * 100
    y = np.random.rand(num_points) * 100
    z = np.random.rand(num_points) * 100
    df = pd.DataFrame({'x': x, 'y': y, 'z': z})
    return df
# plot 3d scatter plot
def plot_3d_scatter(df):
    #install mpl_toolkits library in this environment
    try:
        from mpl_toolkits.mplot3d import Axes3D
    except ImportError:
        !pip install mpl_toolkits
        from mpl_toolkits.mplot3d import Axes3D
        return

    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['x'], df['y'], df['z'], c='blue', marker='o')
    ax.set_title('3D Scatter Plot')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    plt.show()
plot_3d_scatter(generate_3d_scatter_df())
```
